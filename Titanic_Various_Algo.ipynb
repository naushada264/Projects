```python
import pandas as pd
import numpy as np
res = pd.read_csv("../input/titanic/test.csv")
data = pd.read_csv("../input/titanic/train.csv")
```

```python
data[:10]
```

```python
data = data.drop(["Ticket","Fare","Cabin","Embarked","Parch","SibSp"],axis = 1)
res = res.drop(["Ticket","Fare","Cabin","Embarked","Parch","SibSp"],axis = 1)
```

```python
data.fillna(data.Age.mean(), inplace = True)
res.fillna(res.Age.mean(), inplace = True)
```

```python
gender_data = pd.get_dummies(data["Sex"], drop_first = True)
gender_res = pd.get_dummies(res["Sex"], drop_first = True)
```

```python
pclass_data = pd.get_dummies(data["Pclass"], drop_first = True)
pclass_res = pd.get_dummies(res["Pclass"], drop_first = True)
```

```python
data = pd.concat([data,gender_data,pclass_data], axis = 1)
res = pd.concat([res,gender_res,pclass_res], axis = 1)
```

```python
data.drop(["Name","Sex","Pclass"], axis = 1, inplace = True)
res.drop(["Name","Sex","Pclass"], axis = 1, inplace = True)
```

```python
print(data.shape)
print(res.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

models = [
    ('LR', LogisticRegression()),
    ('NB', GaussianNB()),
    ('SVM', SVC()),
    ('KNN', KNeighborsClassifier()),
    ('DT', DecisionTreeClassifier()),
]
```

```python
X = data.drop(["Survived"], axis = 1)
y = data["Survived"]
```

```python
for name, model in models:
    clf = model
    clf.fit(X, y)
    y_pred = clf.predict(res)
    y_data = {'Survived': y_pred}
    new_data = pd.DataFrame(data = y_data, index = res["PassengerId"])
    new_data.to_csv("new_result{}.csv".format(name))
```
